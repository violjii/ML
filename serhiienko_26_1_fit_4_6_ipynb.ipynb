```python
# Лабораторна робота №1
# Завдання 1: Створіть одновимірний масив (вектор) з оберненими значеннями факторіалів натуральних чисел, починаючи з 1, та виведіть його значення.

import numpy as np
from math import factorial

# Обернені значення факторіалів чисел від 1 до 6
factorial_inverses = np.array([1 / factorial(i) for i in range(1, 7)])
factorial_inverses

# Завдання 2: Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його квадратами факторіалів кожного числа, починаючи з 1, та виведіть його значення.

factorial_squares = np.array([[factorial(i)**2 for i in range(1 + j*3, 4 + j*3)] for j in range(3)])
factorial_squares

# Завдання 3: Створіть масив розміром 3x3, заповніть його випадковими натуральними числами від 1 до 10 та знайдіть мін елемент.

random_matrix = np.random.randint(1, 11, (3, 3))
min_element = np.min(random_matrix)
random_matrix, min_element

# Завдання 4: Створіть масив розміром 2x2, заповніть його випадковими дійсними числами від -1.2 до 1.2 та виведіть його значення.

random_float_matrix = np.random.uniform(-1.2, 1.2, (2, 2))
random_float_matrix

# Завдання 5: Створіть два одновимірних масиви розміром 5, заповніть їх випадковими цілими числами в діапазоні від 20 до 30 та виконайте на них поелементні операції додавання, віднімання та множення.

array1 = np.random.randint(20, 31, 5)
array2 = np.random.randint(20, 31, 5)
sum_arrays = array1 + array2
sub_arrays = array1 - array2
mul_arrays = array1 * array2
array1, array2, sum_arrays, sub_arrays, mul_arrays

# Завдання 6: Створіть дві матриці 4x4, заповніть їх довільними цілими числами від 10 до 100 та знайдіть їх добуток.

matrix1 = np.random.randint(10, 101, (4, 4))
matrix2 = np.random.randint(10, 101, (4, 4))
matrix_product = np.dot(matrix1, matrix2)
matrix1, matrix2, matrix_product

# Завдання 7: Створіть дві матриці розміром 3x2 та 2x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.

matrix_3x2 = np.random.randint(1, 6, (3, 2))
matrix_2x4 = np.random.randint(1, 6, (2, 4))
matrix_mult = np.dot(matrix_3x2, matrix_2x4)
matrix_3x2, matrix_2x4, matrix_mult

# Завдання 8: Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 12 та знайдіть її обернену матрицю.

matrix_3x3 = np.random.randint(1, 13, (3, 3))
inverse_matrix = np.linalg.inv(matrix_3x3)
matrix_3x3, inverse_matrix

# Завдання 9: Створіть матрицю розміром 4x4, заповніть її випадковими цілими числами в діапазоні від 0 до 1 та транспонуйте її.

binary_matrix = np.random.randint(0, 2, (4, 4))
transposed_matrix = np.transpose(binary_matrix)
binary_matrix, transposed_matrix

# Завдання 10: Створіть матрицю розміром 3x4 та вектор розміром 4x1, заповніть матрицю випадковими цілими числами в діапазоні від 1 до 10, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть добуток матриці на вектор.

matrix_3x4 = np.random.randint(1, 11, (3, 4))
vector_4x1 = np.random.uniform(0, 1, (4, 1))
matrix_vector_mult = np.dot(matrix_3x4, vector_4x1)
matrix_3x4, vector_4x1, matrix_vector_mult

# Завдання 11: Створіть матрицю розміром 2x3 та вектор розміром 3x1, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими цілими числами в діапазоні від 1 до 10, і знайдіть добуток матриці на вектор.

matrix_2x3 = np.random.uniform(0, 1, (2, 3))
vector_3x1 = np.random.randint(1, 11, (3, 1))
matrix_vector_product = np.dot(matrix_2x3, vector_3x1)
matrix_2x3, vector_3x1, matrix_vector_product

# Завдання 12: Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 3 та знайдіть добуток всіх елементів обох матриць.

matrix_a = np.random.randint(1, 4, (3, 3))
matrix_b = np.random.randint(1, 4, (3, 3))
product_of_elements = np.prod(matrix_a) * np.prod(matrix_b)
matrix_a, matrix_b, product_of_elements

# Завдання 13: Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від -5 до 5 та знайдіть добуток їх елементів.

matrix_c = np.random.randint(-5, 6, (3, 3))
matrix_d = np.random.randint(-5, 6, (3, 3))
element_wise_product = np.prod(matrix_c) * np.prod(matrix_d)
matrix_c, matrix_d, element_wise_product

# Завдання 14: Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 10 та знайдіть середнє значення елементів на головній діагоналі.

matrix_diag = np.random.randint(1, 11, (3, 3))
mean_diagonal = np.mean(np.diag(matrix_diag))
matrix_diag, mean_diagonal

# Завдання 15: Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить найменший елемент у кожному стовпці матриці.

random_real_matrix = np.random.uniform(-1, 1, (3, 3))
min_column_vector = np.min(random_real_matrix, axis=0)
random_real_matrix, min_column_vector

# Висновок: У цій лабораторній роботі ми розглянули створення та маніпуляції з масивами та матрицями за допомогою бібліотеки NumPy, включаючи випадкові значення, арифметичні операції та лінійну алгебру.
```
